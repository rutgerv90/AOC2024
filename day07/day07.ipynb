## --- Part Two ---

```python
import math
import re
import itertools


file = 'input.txt'
# file = 'sample.txt'

# init vars
test_results = []
test_items = []
total_score = 0

# open file
with open(file, 'r') as f:
    lines = f.read().splitlines()
    for line in lines:
        # regex match the line
        items = re.match(r"^(\d+):\s([\d\s]+)$",line)
        # get the test value
        test_value = int(items.group(1))
        # get the remaining values and put them into a list
        remaining = [int(x) for x in items.group(2).split(' ')]
        # append to lists
        test_results.append(test_value)
        test_items.append(remaining)


operators = ['*', '+', '||']

# function to extract () parts of each expression
def extract_parts(expression):
    parts = []
    while '(' in expression:
        # Find the innermost parentheses
        innermost_parentheses = re.search(r'\([^()]*\)', expression)
        if not innermost_parentheses:
            break
        
        # Extract the innermost expression
        inner_expr = innermost_parentheses.group(0)
        
        # Add the innermost expression to the parts list
        parts.append(inner_expr)
        
        # Replace the innermost expression with a placeholder in the original expression
        expression = expression[:innermost_parentheses.start()] + 'X' + expression[innermost_parentheses.end():]
    
    return parts

for i in range(len(test_results)):
    # how many spaces are there to put the '+' and '*'
    spaces = len(test_items[i])-1
    possible = [list(item) for item in list(itertools.product(operators, repeat=spaces))]
    
    # Calculate the percentage
    percent_complete = (i + 1) / len(test_items) * 100
    # Print the progress
    print(f"Loading... {percent_complete:.2f}%", end='\r')

    for item in possible:
        # prepend parentheses to preserve order
        result = []
        # loop over items and add number and then the operator
        for j in range(len(test_items[i]) - 1):
            result.append(str(test_items[i][j]))
            result.append(item[j])
        # add the final number
        result.append(str(test_items[i][-1]))
        # cast to string (used for eval)
        compute_string = ''.join(result)
        # to preserve order add parentheses after every number
        compute_string =re.sub(r"(\d+)",r"\1)",compute_string)
        prepend_nr = len(re.findall(r"\)",compute_string))
        compute_string = '('*prepend_nr +compute_string
        parts = extract_parts(compute_string)

        total = 0
        for part in parts:
            if 'X' in part:
                aaa = re.sub(r"X",str(total), part)
                if '||' in aaa:
                    aaa = re.sub(r"\|\|","",aaa)
                total=eval(aaa)
            else:
                total=eval(part)
            if total > test_results[i]:
                break
        if int(total) == test_results[i]:
            total_score += test_results[i]
            # only count 1 result per eval
            break

print('\nAnswer to part 2:', total_score)
```
