## --- Part Two ---

```python
# precompute the length for all base 0-9 numbers for a manageable number of blinks; 40 seemed to work
from collections import Counter

future = dict()
blinks = 40 # manageable number

# get all nrs from 0 to 9
for i in range(10):

    # ---- checking and displaying for loop loading times ----
    # Calculate the percentage
    percent_complete = (i + 1) / 10 * 100
    # Print the progress
    # print(f"Part {i+1} Loading... {percent_complete:.2f}%", end='\r')
    # --------------------------------------------------------

    hold = Counter()
    test = [i]
    for j in range(blinks):

        updated = []

        for k in range(len(test)):
            item = test[k]
            str_item = str(item)
            # if 0 set to 1
            if item == 0:
                updated.append(1)
            # if even split into 2 parts
            elif len(str(item)) % 2 ==0:
                part1 = int(str_item[:int((len(str_item)/2))])
                part2 = int(str_item[int((len(str_item)/2)):])
                updated.append(part1)
                updated.append(part2)
            # else multiply item by 2024
            else:
                updated.append(item*2024)
        test = updated
        # print(i, j , updated)
        hold[j+1] = len(test)
    future[i] = hold
    # print('---')
```
