## --- Part One ---

```python
import numpy as np
from copy import deepcopy

file = 'input.txt'
# file = 'sample.txt'
# file = 'sample2.txt'

# init vars
total_score = 0
input_string = ''
result = []

# open file & load content
with open(file, 'r') as f:
    input_string = f.read()

new_index = 0
for x in range(len(input_string)):
    # add the right number of '.'
    if x%2 == 1:
        result.extend('.'*int(input_string[x]))
    else:
        # add the index numbers
        for j in range(int(input_string[x])):
            result.append(str(new_index))
        # up the overall index
        new_index+=1

# find every index where there's a '.'
open_spaces = np.where(np.array(result) =='.')[0].tolist()

# loop over the results in reverse order
for x,item in enumerate(reversed(result)):
    # keep track of the reverse index
    r_idx = len(result)-x-1
    # if the amount of open spaces left == 0 or the next open space is past the reverse index break the loop
    if len(open_spaces) == 0 or open_spaces[0] > r_idx:
        break
    # if item is a number; pop the next open space and put the number there and replace the current space with a '.'
    if item != '.':
        x = open_spaces.pop(0)
        result[x] = item
        result[r_idx] = '.'

# remove all the '.'
test = [item for item in result if item != '.']

# loop over the results and calculate the score
for x,item in enumerate(test):
    total_score += (x*int(item))

print('Answer to part 1:', total_score)
```

## --- Part Two ---

```python
import numpy as np
from copy import deepcopy

file = 'input.txt'
# file = 'sample.txt' 
# file = 'sample2.txt'

# init vars
total_score = 0
input_string = ''
result = []
open_spaces = []

# open file & load content
with open(file, 'r') as f:
    input_string = f.read()

new_index = 0
# slightly different, we put everything in a list now immediately
for x in range(len(input_string)):
    # add the right number of '.'
    if x%2 == 1:
        if int(input_string[x]) > 0:
            result.append(list('.'*int(input_string[x])))
    else:
        score_index = []
        # add the index numbers
        for j in range(int(input_string[x])):
            score_index.append(str(new_index))
        result.append(score_index)
        # up the overall index
        new_index+=1

# loop over the results
for y in range(len(result)):

    # ---- checking and displaying for loop loading times ----
    # Calculate the percentage
    percent_complete = (y + 1) / len(result) * 100
    # Print the progress
    print(f"Loading... {percent_complete:.2f}%", end='\r')
    # --------------------------------------------------------

    # find the reverse index
    r_idx = len(result)-y-1
    # grab the item
    item = result[r_idx]
    # check if item is not '.' and len > 0
    if len(item) > 0 and item[0] != '.':
        # set the block size (reserved space) to be the length of the item
        block_size = len(item)
        # loop over the results in order 
        for j, space in enumerate(result):
            # if the reverse index is smaller than the index of the loop break out
            if r_idx <= j:
                break
            if len(space) > 0 and space[0] == '.':
                # suitable space is found
                if len(space) >= block_size:
                    # set a temporary new reverse index, since we will be changing list size
                    r_idx2 = r_idx
                    # if no space is left over, remove the item
                    if space[block_size:] == []:
                        # shrinks size by one
                        result.pop(j)
                        # since we removed 1 element, decrease the reverse index by one
                        r_idx2 = r_idx-1
                    else:
                        # replace the space with the leftover space (if exist)
                        result[j] = space[block_size:]
                    # replace item with the replace space
                    result[r_idx2] = space[:block_size]
                    # insert the item into the new area
                    result.insert(j,item)
                    break

score_index = 0
# loop over the results and calculate the score
for item in result:
    # loop over all items in a position
    for y,sub in enumerate(item):
        # only calculate if its a number
        if sub != '.':
            total_score += score_index*int(sub)
        # up the index by 1
        score_index +=1


print('\nAnswer to part 2:', total_score)
```
