```python
import numpy as np
from pprint import pprint
import re
import math
import sys

# dont truncate numpy output
np.set_printoptions(threshold=sys.maxsize)

file = 'input.txt'
# file = 'sample.txt'
# file = 'sample2.txt'

# init vars
total_score = 0
warehouse = []
directions = []

# open file & load content
with open(file, 'r') as f:
    [room, instructions] = f.read().split('\n\n')
    for line in room.splitlines():
        warehouse.append(list(line))
    for line in instructions.splitlines():
        directions.append(list(line))

# convert to np array
warehouse = np.array(warehouse)
# flatten directions 
directions = np.array(directions).flatten()

# find start location  
start = np.where(warehouse == '@',)

# set robot location [row,col]
robot = [start[0][0],start[1][0]]

def print_maze(array):
    for a in array:
        print(''.join(a))

# start moving 
for i in range(len(directions)):
    direction = directions[i]
    moved = False
    if direction == '<':
        spaces = warehouse[robot[0]][0:robot[1]]
        r_spaces = np.flip(spaces)
        for idx,j in enumerate(r_spaces):
            if j == '#':
                break
            if j == '.':
                # move everything before it 1 left
                r_spaces = np.delete(r_spaces, idx)
                # include the new robot location
                r_spaces = np.insert(r_spaces,0, '@')
                warehouse[robot[0]][0:robot[1]] = np.flip(r_spaces)
                # set previous robot loc 
                warehouse[robot[0], robot[1]] = '.'
                # move new robot loc 1 left
                robot[1] -=1
                break
    elif direction == '>':
        spaces = warehouse[robot[0]][robot[1]+1:]
        for idx,j in enumerate(spaces):
            if j == '#':
                break
            if j == '.':
                # move everything before it 1 right
                spaces = np.delete(spaces, idx)
                # include the new robot location
                spaces = np.insert(spaces,0, '@')
                warehouse[robot[0]][robot[1]+1:] = spaces
                # set previous robot loc 
                warehouse[robot[0], robot[1]] = '.'
                # move new robot loc 1 right
                robot[1] +=1
                break
    elif direction == '^':
        spaces = warehouse[:robot[0],robot[1]]
        r_spaces = np.flip(spaces)
        for idx,j in enumerate(r_spaces):
            if j == '#':
                break
            if j == '.':
                # move everything before it 1 left
                r_spaces = np.delete(r_spaces, idx)
                # include the new robot location
                r_spaces = np.insert(r_spaces,0, '@')
                warehouse[:robot[0],robot[1]] = np.flip(r_spaces)
                # set previous robot loc 
                warehouse[robot[0], robot[1]] = '.'
                # move new robot loc 1 up
                robot[0] -=1
                break
    elif direction == 'v':
        spaces = warehouse[robot[0]+1:,robot[1]]
        for idx,j in enumerate(spaces):
            if j == '#':
                break
            if j == '.':
                # move everything before it 1 right
                spaces = np.delete(spaces, idx)
                # include the new robot location
                spaces = np.insert(spaces,0, '@')
                warehouse[robot[0]+1:,robot[1]] = spaces
                # set previous robot loc 
                warehouse[robot[0], robot[1]] = '.'
                # move new robot loc 1 down
                robot[0] +=1
                break

print('final warehouse')
print_maze(warehouse)

boxes = np.where(warehouse=='O')
# calculate total score
for i in range(len(boxes[0])):
    row = boxes[0][i]
    col = boxes[1][i]
    local_score = (row * 100) + col
    total_score += local_score

print('\nAnswer to part 1:', total_score)
```

## --- Part Two ---

```python
import numpy as np
from pprint import pprint
import re
import math
import sys

# dont truncate numpy output
np.set_printoptions(threshold=sys.maxsize)

file = 'input.txt'
# file = 'sample.txt'
# file = 'sample3.txt'

# init vars
total_score = 0
warehouse = []
directions = []

# open file & load content
with open(file, 'r') as f:
    [room, instructions] = f.read().split('\n\n')
    for line in room.splitlines():
        abc = ''
        for x in list(line):
            if x == '#':
                abc += '##'
            elif x == 'O':
                abc += '[]'
            elif x == '.':
                abc += '..'
            elif x == '@':
                abc += '@.'
        warehouse.append(list(abc))
    for line in instructions.splitlines():
        directions.append(list(line))

# convert to np array
warehouse = np.array(warehouse)
# flatten directions 
directions = np.array(directions).flatten()

# find start location  
start = np.where(warehouse == '@',)

# set robot location [row,col]
robot = [start[0][0],start[1][0]]

def print_maze(array):
    for a in array:
        print(''.join(a))

# checks items above location and returns them
def check_above(locations, area):
    can_move = True

    for idx,item in enumerate(locations):
        above_item = area[item[0]-1,item[1]]
        if above_item == ']':
            if [item[0]-1, item[1]-1] not in locations:
                locations.append([item[0]-1, item[1]-1])           
            if [item[0]-1, item[1]] not in locations:
                locations.append([item[0]-1, item[1]])      
        elif above_item == '[':
            if [item[0]-1, item[1]+1] not in locations:
                locations.append([item[0]-1, item[1]+1])           
            if [item[0]-1, item[1]] not in locations:
                locations.append([item[0]-1, item[1]])  
        # if a wall is hit box(es) cant move
        elif above_item == '#':
            can_move = False
            break
    
    return [can_move, locations]

# checks items below location and returns them
def check_below(locations, area):
    can_move = True

    for idx,item in enumerate(locations):
        below_item = area[item[0]+1,item[1]]
        if below_item == ']':
            if [item[0]+1, item[1]-1] not in locations:
                locations.append([item[0]+1, item[1]-1])           
            if [item[0]+1, item[1]] not in locations:
                locations.append([item[0]+1, item[1]])      
        elif below_item == '[':
            if [item[0]+1, item[1]+1] not in locations:
                locations.append([item[0]+1, item[1]+1])           
            if [item[0]+1, item[1]] not in locations:
                locations.append([item[0]+1, item[1]])      
        # if a wall is hit box(es) cant move
        elif below_item == '#':
            can_move = False
            break
    
    return [can_move, locations]


for i in range(len(directions)):
    direction = directions[i]
    # left direction stays the same as in part 1
    if direction == '<':
        spaces = warehouse[robot[0]][0:robot[1]]
        r_spaces = np.flip(spaces)
        for idx,j in enumerate(r_spaces):
            if j == '#':
                break
            if j == '.':
                # move everything before it 1 left
                r_spaces = np.delete(r_spaces, idx)
                # include the new robot location
                r_spaces = np.insert(r_spaces,0, '@')
                warehouse[robot[0]][0:robot[1]] = np.flip(r_spaces)
                # set previous robot loc 
                warehouse[robot[0], robot[1]] = '.'
                # move new robot loc 1 left
                robot[1] -=1
                break
    # right direction stays the same as in part 1
    elif direction == '>':
        spaces = warehouse[robot[0]][robot[1]+1:]
        for idx,j in enumerate(spaces):
            if j == '#':
                break
            if j == '.':
                # move everything before it 1 right
                spaces = np.delete(spaces, idx)
                # include the new robot location
                spaces = np.insert(spaces,0, '@')
                warehouse[robot[0]][robot[1]+1:] = spaces
                # set previous robot loc 
                warehouse[robot[0], robot[1]] = '.'
                # move new robot loc 1 right
                robot[1] +=1
                break
    # up and down direction functionality changed
    elif direction == '^':
        [moveable, moves] = check_above([robot], warehouse)
        if moveable:
            # in reverse order move all of them 1 up
            for item in reversed(moves):
                charx = warehouse[item[0],item[1]]
                # move
                warehouse[item[0]-1,item[1]] = charx
                warehouse[item[0],item[1]] = '.'
            # move robot 1 up
            robot[0] -=1

    # up and down direction functionality changed
    elif direction == 'v':
        [moveable, moves] = check_below([robot], warehouse)
        if moveable:
            # in reverse order move all of them 1 down
            for item in reversed(moves):
                charx = warehouse[item[0],item[1]]
                # move
                warehouse[item[0]+1,item[1]] = charx
                warehouse[item[0],item[1]] = '.'
            # move robot 1 down
            robot[0] +=1

print('final warehouse')
print_maze(warehouse)

boxes = np.where(warehouse=='[')
for i in range(len(boxes[0])):
    row = boxes[0][i]
    col = boxes[1][i]
    local_score = (row * 100) + col
    total_score += local_score

print('\nAnswer to part 2:', total_score)
```
