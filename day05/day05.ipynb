## --- Part One ---

```python
import math
from pprint import pprint

file = 'input.txt'
# file = 'sample.txt'

# init vars
before_after = dict()
updates = []
correct = []
total_score = 0

# open file
with open(file, 'r') as f:
    # read input string, split into two parts
    [page_order,update_order] = [line.splitlines() for line in f.read().split("\n\n")]
    # first do the page ordering
    for line in page_order:
        # separate the before&after
        [before,after] = [int(x) for x in line.split('|')]
        # if it exists append the 'after' to the before list if not create it
        if before in before_after:
            before_after[before].append(after)
        else:
            before_after[before] = [after]
    # convert to ints and append to updates list 
    for line in update_order:
        updates.append([int(x) for x in line.split(',')])

# loop through the updates
for line in updates:
    # keep track of previous numbers
    previous = []
    # number of problems
    problems = 0
    # loop through the pages
    for page in line:
        # check if page exist in the before_after dict
        if page in before_after:
            # check if any of the previous numbers overlap with the dict results
            overlap_count = len(set(previous) & set(before_after[page]))
            problems += overlap_count
        # add page to list
        previous.append(page)
    # only append the line if no problems found
    if problems == 0:
        correct.append(line)

# loop through the correct entries and get the middle number, add to total score
for line in correct:
    # get the middle number
    index = math.floor(len(line)/2)
    # add to total score
    total_score += line[index]

print('Answer to part 1:', total_score)
```

## --- Part Two ---

```python
import math
from pprint import pprint

file = 'input.txt'
# file = 'sample.txt'

# init vars
before_after = dict()
updates = []
correct = []
# add incorrect list
incorrect = []
total_score = 0

# open file
with open(file, 'r') as f:
    # read input string, split into two parts
    [page_order,update_order] = [line.splitlines() for line in f.read().split("\n\n")]
    # first do the page ordering
    for line in page_order:
        # separate the before&after
        [before,after] = [int(x) for x in line.split('|')]
        # if it exists append the 'after' to the before list if not create it
        if before in before_after:
            before_after[before].append(after)
        else:
            before_after[before] = [after]
    # convert to ints and append to updates list 
    for line in update_order:
        updates.append([int(x) for x in line.split(',')])

# loop through the updates
for line in updates:
    # keep track of previous numbers
    previous = []
    # number of problems
    problems = 0
    # loop through the pages
    for page in line:
        # check if page exist in the before_after dict
        if page in before_after:
            # check if any of the previous numbers overlap with the dict results
            overlap_count = len(set(previous) & set(before_after[page]))
            problems += overlap_count
        # add page to list
        previous.append(page)
    # only append the line if problems found
    if problems > 0:
        incorrect.append(line)

# loop through the updates
for line in incorrect:
    # inital check is the original line
    recheck = [line]
    # loop through the recheck list
    for check in recheck:
        # keep track of previous numbers
        previous = []
        # number of problems
        problems = 0
        newlist = []
        # loop through the pages, keep track to index
        for idx,page in enumerate(check):
            # check if page exist in the before_after dict
            if page in before_after:
                # check if any of the previous numbers overlap with the dict results
                overlap_count = len(set(previous) & set(before_after[page]))
                problems += overlap_count
                # if there are problems move the number forward by the number of overlaps in the recheck list
                if overlap_count > 0:
                    newlist.insert(idx-overlap_count, page)
                else:
                    newlist.append(page)
            else:
                newlist.append(page)
            # add page to list
            previous.append(page)
        # only append the line if no problems found
        if problems == 0:
            correct.append(check)
        else:
            # add the new list to the recheck list 
            recheck.append(newlist)


# loop through the correct entries and get the middle number, add to total score
for line in correct:
    # get the middle number
    index = math.floor(len(line)/2)
    # add to total score
    total_score += line[index]

print('Answer to part 2:', total_score)
```
